## Get info box of all movies

```python
#Load web page
r= requests.get("https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films")

#Convert to BSoup object
soup = bs(r.content)

#print the HTML
contents =soup.prettify()
print(contents)
```

```python
# Use select to search if contains CSS classes
movies = soup.select(".wikitable.sortable i")
movies[0:5]
```

```python
#Function to get content value
def get_content(row_data):
  if row_data.find("li"):
    return [li.get_text(" ", strip = True).replace("\xa0", " ") for li in row_data.find_all("li")]
  elif row_data.find("br"):
    return [text for text in row_data.stripped_strings]
  else:
    return row_data.get_text(" ", strip = True).replace("\xa0", " ")

#Remove the references [1],[2]
def remove_tags(soup):
  for tag in soup.find_all(["sup","span"]):
    tag.decompose()


# Get infobox from URL
def get_info_box(url):

  r= requests.get(url)

  #Convert to BSoup object
  soup = bs(r.content)

  # get the infobox vevent
  info_box=soup.find(class_="infobox vevent")
  info_rows= info_box.find_all("tr")

  remove_tags(soup)

  #store rows of html file into dic
  movie_info={}
  for index,row in enumerate(info_rows):
    if index == 0:
      # Get the title
      movie_info["title"] = row.find("th").get_text(" ", strip = True)
    elif index == 1:
      continue
    else: 
      content_key = row.find("th").get_text(" ", strip = True)
      content_value = get_content(row.find("td"))
      movie_info[content_key] = content_value

  return movie_info
```

```python
#Convert the running time into interger

def convert_running_time(running_time):
  if running_time == "N/A":
    return None

  #["95 min","75 min"]
  if isinstance(running_time, list):
    first_element = running_time[0]
    value = int(first_element.split(" ")[0])
  #["74 mins"]
  else:
    value = int(running_time.split(" ")[0])
  return value

for movie in movies_info_list:
  movie["Running time INT"] = convert_running_time(movie.get("Running time","N/A"))
```

## Convert Budget and Box Office to float

```python
#Convert Budget and Box Office to float
[movie.get("Budget", "N/A") for movie in movies_info_list]
```

```python
import re

amounts = r"thousand|million|billion"
number = r"\d+(,\d{3})*\.*\d*"
standard = fr"\${number}(-|\sto\s)?({number})?\s({amounts})"

#Convert million,billion to interger
def word_to_value(word):
	value_dict = {"thousand": 1000, "million": 1000000, "billion": 1000000000}
	return value_dict.get(word.lower(), 1)

# Convert xx millions to interger
def parse_word_syntax(string):
	stripped_string = string.replace(",", "")
	value = float(re.search(number, stripped_string).group())
	modifier = word_to_value(re.search(amounts, string, flags=re.I).group())
	return value*modifier

def parse_value_syntax(string):
	stripped_string = string.replace(",", "")
	return float(re.search(number, stripped_string).group())

def money_conversion(money):
	if type(money) == list:
		money = money[0]

	word_syntax = re.search(standard, money, flags=re.I)
	value_syntax = re.search(fr"\${number}", money)

	if word_syntax:
		return parse_word_syntax(word_syntax.group())
	elif value_syntax:
		return parse_value_syntax(value_syntax.group())
	else:
		return None
```

```python
for movie in movies_info_list:
  movie["Budget FLOAT"] = money_conversion(movie.get("Budget","N/A"))
  movie["Box office FLOAT"] = money_conversion(movie.get("Box office","N/A"))
```

## Convert Dates into datetime objects

```python
[movie.get("Release date", "N/A") for movie in movies_info_list]
```

```python
from datetime import datetime
dates = [movie.get("Release date", "N/A") for movie in movies_info_list]

#Split the date from location ---May 7, 2011 ( Disneyland Resort )
def clean_date(date):
  return date.split("(")[0].strip()

def date_convert(date_1):
  if isinstance(date_1,list):
    date_1 = date_1[0]

  if date_1 == "N/A":
    return None

  date_str = clean_date(date_1)
  # convert diffrent format ----- October 23, 2012 --- 26 October 2008
  fmts=["%B %d, %Y", "%d %B, %Y"]
  for fmt in fmts:
    try:
      return datetime.strptime(date_str,fmt)
    except:
      pass
  return None
```

```python
for movie in movies_info_list:
  if 'Release dates' in movie.keys():# multiple release dates
    movie["Release date DATETIME"] = date_convert(movie.get("Release dates", "N/A"))
  else:# only 1 release day
    movie["Release date DATETIME"] = date_convert(movie.get("Release date", "N/A"))
```
