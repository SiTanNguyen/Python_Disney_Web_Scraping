# Explore Data Analysis

## Question 1: How many films were created in the 20 and 21 century? Average Imdb, Metascore , Rotten tomatoes of each century

```python
df["Year release"] = pd.DatetimeIndex(df['Release date DATETIME']).year
```

```python
century_df = df[['title', 'Release date DATETIME', 'Year release', 'Imdb', 'Metascore', 'Rotten tomato score']]
century_df
```

```python
#Remove the NaN and change type of year release
century_df = century_df.fillna(0)
century_df["Year release"].astype(int)

def get_century(year):
  if year in range(1900,2000):
    return "20 century"
  elif year in range(2000,3000):
    return "21 century"
  else: 
    return "None listed"

century_df['Century created'] = century_df['Year release'].apply(lambda x: get_century(x))
```

```python
century_df[century_df["Century created"] == "21 century"]
```

```python
century_df_group = century_df.groupby("Century created").count()
century_df_group["title"]
```

```python
century_score_group = century_df.groupby("Century created").mean()
century_score_group
```

##Question 4: What are the highest profit movies

```python
profit_df = df[['title', "Directed by" ,'Release date DATETIME', 'Imdb', "Budget FLOAT", "Box office FLOAT"]]
profit_df
```

```python
profit_df = profit_df.fillna(0)
profit_df["profit"] = (profit_df['Box office FLOAT'] - profit_df['Budget FLOAT'])
profit_df["ROI (%)"] = round((profit_df['Box office FLOAT'] - profit_df['Budget FLOAT'])/ profit_df['Budget FLOAT'],1)
profit_df
```

```python
profit_df.sort_values("profit", ascending = False)
```
